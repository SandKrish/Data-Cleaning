```python
result_df = result_df.copy()
```

```python
result_df['Video_Title'] = result_df['Video_Title'].replace(r'\n',' ', regex=True) 
result_df['Video_Title'].str.strip()
result_df.head(2)
```

```python
display(result_df[result_df['Channel_Name'].str.contains('\n', na= False)])
display(result_df[result_df['Total_Views'].str.contains('\n', na = False)])
display(result_df[result_df['Publish_Date'].str.contains('\n', na = False)])
display(result_df[result_df['Description'].str.contains('\n', na = False)])
```
